## Pasos previos ##

### Uso de xplico ###

1. Corra los sevicios necesarios

```
sudo service apache2 start
sudo service xplico start
```

2. Acceda en el browser:

```
http://localhost:9876/
```

3. Logeese:
   * **Usuario normal**: u --> xplico; p --> xplico
   * **Usuario poderoso**: u --> admin; p --> xplico
   
   
### Instalacion y uso de capanalysis ###

1. Se arranca la aplicación:

```
sudo /etc/init.d/capanalysis restart
```

2. Lleve a cabo los analisis en: ```http://localhost:9877/datasets```

## Analisis desde capanalysis ##
Para este caso el dataset asociado con la traza **synflood.pcap** se nombro como **synfloodtraza**, el primer analisis al respecto se muestra a continuacion:

Haciendo un analisis superficial tenemos:
1. **Flujos**: 
   * Tamaño: 57.7 k ???
   * IP-src (2 fuentes): 192.168.0.100, fe80::7ed1::c3ff::fe94::9eb8
   * IP-dst: (3 destinos) 192.168.0.101, 224.0.0.251, ff02::fb
   * Puertos fuente: Aleatorios.
   * Puertos destino: 80.
   * L4 = TCP, L3 = HTTP.
   * No se enviaron ni se recibieron bytes.
   * Flujos HTTP: 65646 (aprox 65.6 k).
   * Solo se enviaron datos MDNS: 263 B
   * No se recibio ningun dato.
   * Se hacen graficas de source IP .vs. flujos: Se muestra que se envian aproximadamente 65.6 k, casi todos Http. La gran mayoria de los flujos (casi los 65 k) tienen una duración de 1.5 Ms.
2. **Estadisticas por hora**: El tiempo de captura se dio el dia 2014-02-25 entre las 0:00 y 0:59.
3. **Resumen sobre los flujos**:
   * Http:
     * IP src: 1
     * IP dst: 1
     * Data: 0
     * Flujos: 65.6 k
   * Http:
     * IP src: 2
     * IP dst: 2
     * Data: 268 B
     * Flujos: 2
   
4. En el caso en el cual hay muchos flujos se puede notar un gran numero de conexiones (Ver detalla de las direcciones IP afectadas), probablemento esto es lo que muestra los estragos del ataque de sy floofind.

The IO graph statistics show the following summary:
The TCP connection never closes as there is no count for tcp.flags.fin
The TCP connection never exchanges any data as there is no count for
tcp.flags.push
The count of SYN packets is very high
The count of ACK is half of that of the SYN packets
In real scenarios, this data will be mixed up with actual packet flows, but the analysis
technique will remain the same. The moment you see an unexpected growth in SYN
packets or a spike in SYN packets, it’s a SYN flood from DoS or from the multiple-source
DDoS.
